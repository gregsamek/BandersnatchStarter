```python
def chart(df: DataFrame, x: str, y: str, target: str) -> Chart:
    graph = Chart(
        df,
        title=f"{y} by {x} for {target}",
    ).mark_circle(size=100).encode(
        x=x,
        y=y,
        color=target,
        tooltip=Tooltip(df.columns.to_list())
    )
    return graph
```

```python
chart(df, 'Type', 'Energy', 'Health')
```

```python
def chart(df: DataFrame, x: str, y: str, target: str) -> Chart:
    graph = Chart(
        df,
        title=f"{y} by {x} for {target}"
    ).mark_circle(size=100).encode(
        x=x,
        y=y,
        color=target,
        tooltip=Tooltip(df.columns.to_list())
    ).properties(
        width=500, # set width of chart
        height=400 # set height of chart
    ).interactive() # make chart interactive
    return graph

chart(df, 'Type', 'Energy', 'Health')
```

```python
def chart(df: DataFrame, x: str, y: str, target: str) -> Chart:
    return Chart(
        df, 
        title=f"{y} by {x} for {target}",
        padding={"left": 10, "top": 10, "right": 10, "bottom": 10}
    ).mark_circle(size=100).encode(
        x=x,
        y=y,
        color=target,
        tooltip=Tooltip(df.columns.to_list()),
    ).configure_title(
        fontSize=20, # Set font size for chart title
    ).configure_axis(
        labelFontSize=14,  # Set font size for axis labels
        titleFontSize=14  # Set font size for axis titles
    ).configure_legend(
        labelFontSize=12,
        titleFontSize=12  # Set font size for legend title
    ).properties(
        width=500,
        height=400
    ).interactive()

chart(df, 'Type', 'Energy', 'Health')
```

```python
def chart(df: DataFrame, x: str, y: str, target: str) -> Chart:
    return Chart(
        df, 
        title=f"{y} by {x} for {target}",
        background="#252525", # set color
        padding={"left": 10, "top": 10, "right": 10, "bottom": 10}
    ).mark_circle(size=100).encode(
        x=x,
        y=y,
        color=target,
        tooltip=Tooltip(df.columns.to_list()),
    ).configure_title(
        fontSize=20,
        color="#aaaaaa"  # set color
    ).configure_axis(
        labelColor="#aaaaaa", # set color
        titleColor="#aaaaaa", # set color
        labelFontSize=14,
        titleFontSize=14
    ).configure_legend(
        labelColor="#aaaaaa", # set color
        titleColor="#aaaaaa", # set color
        labelFontSize=12,
        titleFontSize=12
    ).properties(
        width=500,
        height=400
    ).interactive()

chart(df, 'Type', 'Energy', 'Health')
```
